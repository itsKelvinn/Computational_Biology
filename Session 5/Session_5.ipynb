```python
CreatedFastaFile = sio.parse('sesi5fasta.fasta', 'fasta')
a = next(CreatedFastaFile)

SeqA = a.seq
Seq1 = SeqA[:round(len(SeqA) / 2)]
Seq2 = SeqA[round(len(SeqA) / 2):]

print(Seq1)
print(Seq2)
```

```python
from Bio.SeqUtils import gc_fraction as gc, MeltingTemp as mt
```

```python
gc1 = gc(Seq1)
gc2 = gc(Seq2)

mt1 = mt.Tm_NN(Seq1, strict=False)
mt2 = mt.Tm_NN(Seq2, strict=False)

print(gc1)
print(gc2)
print(mt1)
print(mt2)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axis = plt.subplots(1, 2)
axis[0].bar(["GC Seq1", "GC Seq2"], [gc1, gc2])
axis[1].bar(["MT Seq", "MT Seq2"], [mt1, mt2])

axis[0].set_title("GC Percentage")
axis[1].set_title("Melting Temp")

plt.show()
```
